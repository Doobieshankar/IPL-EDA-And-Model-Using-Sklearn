# IPL Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# display settings
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")
```

```python
# Load CSV file
df = pd.read_csv("IPL.csv")  # change filename

# Preview data
df.head()
```

```python
print("Shape:", df.shape)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.describe(include="all")
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
df.fillna({
    'extra_type':'none',
    'wicket_count':'none',
    'player_out':'none'
           },inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
team_runs = df.groupby('batting_team')['runs_total'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
team_runs.plot(kind='bar')
plt.title("Total Runs by Team")
plt.xlabel("Team")
plt.ylabel("Runs")
plt.xticks(rotation=45)
plt.show()
```

```python
top_batters = df.groupby('batter')['runs_batter'].sum().sort_values(ascending=False).head(10)
plt.Figure(figsize=(10,5))
sns.barplot(x=top_batters.values, y=top_batters.index)
plt.title("Top 10 Batters")
plt.xlabel("Runs")
plt.ylabel("Batter")
plt.show()
```

```python
wickets = df[df['wicket_kind'] != 'None']

top_bowlers = wickets['bowler'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_bowlers.values, y=top_bowlers.index)
plt.title("Top 10 Bowlers (Wickets)")
plt.xlabel("Wickets")
plt.ylabel("Bowler")
plt.show()
```

```python
wins = df[['match_won_by']].dropna()

plt.figure(figsize=(10,5))
sns.countplot(y='match_won_by', data=wins, order=wins['match_won_by'].value_counts().index)
plt.title("Match Wins by Team")
plt.xlabel("Count")
plt.ylabel("Team")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='toss_decision', data=df)
plt.title("Toss Decision Distribution")
plt.show()
```

```python
runs_per_over = df.groupby('over')['runs_total'].sum()

plt.figure(figsize=(10,5))
runs_per_over.plot()
plt.title("Runs per Over")
plt.xlabel("Over")
plt.ylabel("Runs")
plt.show()
```

```python
venue_runs = df.groupby('venue')['runs_total'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=venue_runs.values, y=venue_runs.index)
plt.title("Top Venues by Runs")
plt.xlabel("Runs")
plt.ylabel("Venue")
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=np.number)
plt.figure(figsize=(12,8))
sns.heatmap(numeric_cols.corr(), cmap='coolwarm',annot=False)
plt.title("correlation Heatmap")
plt.show()
```

```python
plt.Figure(figsize=(8,5))
sns.countplot(x='extra_type', data=df)
plt.title("Extras Type  Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
year_matches = df['year'].value_counts().sort_index()

plt.figure(figsize=(8,5))
year_matches.plot(kind='bar')
plt.title("Matches per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
    # Strike Rate = (runs / balls) * 100
batter_stats = df.groupby('batter').agg({
    'runs_batter': 'sum',
    'balls_faced': 'sum'
})

batter_stats['strike_rate'] = (batter_stats['runs_batter'] / batter_stats['balls_faced']) * 100
top_sr = batter_stats.sort_values(by='strike_rate', ascending=False).head(10)

top_sr
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=top_sr['strike_rate'], y=top_sr.index)
plt.title("Top Batters by Strike Rate")
plt.show()
```

```python
# Economy = runs given / overs
bowler_stats = df.groupby('bowler').agg({
    'runs_bowler': 'sum',
    'valid_ball': 'sum'
})

bowler_stats['overs'] = bowler_stats['valid_ball'] / 6
bowler_stats['economy'] = bowler_stats['runs_bowler'] / bowler_stats['overs']

best_economy = bowler_stats.sort_values(by='economy').head(10)
best_economy
```

```python
# Assuming runs_batter contains boundary info
boundaries = df[df['runs_batter'].isin([4,6])]

boundary_counts = boundaries['runs_batter'].value_counts()

plt.figure(figsize=(6,4))
sns.barplot(x=boundary_counts.index, y=boundary_counts.values)
plt.title("Boundary Distribution (4s vs 6s)")
plt.xlabel("Runs")
plt.ylabel("Count")
plt.show()
```

```python
def phase(over):
    if over <= 6:
        return 'Powerplay'
    elif over <= 15:
        return 'Middle'
    else:
        return 'Death'

df['phase'] = df['over'].apply(phase)

phase_runs = df.groupby('phase')['runs_total'].sum()

phase_runs.plot(kind='bar', figsize=(8,5), title="Runs by Match Phase")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(y='wicket_kind', data=df, order=df['wicket_kind'].value_counts().index)
plt.title("Wicket Types Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['runs_total'], bins=20, kde=True)
plt.title("Runs Distribution per Ball")
plt.show()
```

```python
match_id_sample = df['match_id'].iloc[0]

match_df = df[df['match_id'] == match_id_sample]

match_df['cumulative_runs'] = match_df['runs_total'].cumsum()

plt.figure(figsize=(10,5))
plt.plot(match_df['ball_no'], match_df['cumulative_runs'])
plt.title("Cumulative Runs in Match")
plt.xlabel("Ball Number")
plt.ylabel("Runs")
plt.show()
```

```python
partnerships = df.groupby('batting_partners')['runs_total'].sum().sort_values(ascending=False).head(10)

partnerships
```

```python
toss_vs_win = df[['toss_winner', 'match_won_by']].dropna()

toss_win_match = (toss_vs_win['toss_winner'] == toss_vs_win['match_won_by']).value_counts()

plt.figure(figsize=(6,4))
sns.barplot(x=toss_win_match.index, y=toss_win_match.values)
plt.title("Did Toss Winner Also Win Match?")
plt.xticks([0,1], ['No','Yes'])
plt.show()
```

```python
wickets_df = df[df['wicket_kind'] != 'None']

wickets_per_over = wickets_df.groupby('over').size()

plt.figure(figsize=(10,5))
wickets_per_over.plot()
plt.title("Wickets per Over")
plt.show()
```

```python
pivot = df.pivot_table(values='runs_total', index='over', columns='batting_team', aggfunc='sum')

plt.figure(figsize=(12,8))
sns.heatmap(pivot, cmap='coolwarm')
plt.title("Runs Heatmap (Over vs Team)")
plt.show()
```

```python
pom = df['player_of_match'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=pom.values, y=pom.index)
plt.title("Top Player of the Match Winners")
plt.show()
```
